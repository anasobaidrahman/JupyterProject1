```python
import pandas as pd
import numpy as np
```

```python
data = np.array([[23,51.5],
                 [24,54.5],
                 [39,57.5],
                 [46,60.5],
                 [54,63.5],
                 [66,66.6],
                 [49,69.5],
                 [38,72.5],
                 [21,75.5],
                 [12,78.5]])
df = pd.DataFrame(data, columns=['f','x'])
df
```

```python
df['fx'] = df['f'] * df['x']
df
```

```python
df['c']=df['f'].cumsum()
df
```

```python
df['fx2'] = df['f'] * (df['x']**2)
df
```

```python
df['f|x-x̄|'] = df['f'] * abs(df['x'] - (df['fx'].sum() / df['f'].sum()))
df
```

```python
varience = (df['fx2'].sum() / df['f'].sum()) - (df['fx'].sum() / df['f'].sum())**2
varience
```

```python
Mean_deviation_from_mean = df['f|x-x̄|'].sum() / df['f'].sum()
Mean_deviation_from_mean
```
