```python
salles = np.random.randint(30,size=(4,4))
slles_df = pd.DataFrame(salles,index=['mon','tue','wed','thu'],columns=['item_1','item_2','item_3','item_4'])
```

```python
price= np.random.randint(100,size=(1,4))
price_df = pd.DataFrame(price, index=['price'], columns=['item_1','item_2','item_3','item_4'])
price_df
```

```python
total_sales = salles @ price.T
total_sales
```

```python
salles_df = pd.DataFrame(total_sales, columns=['total_sales'], index=['mon','tue','wed','thu'])
salles_df
```

```python
slles_df['total_sales'] = salles_df
slles_df = pd.concat([slles_df, price_df])
```

```python
slles_df.loc['price','total_sales']=slles_df['total_sales'].sum()
```
