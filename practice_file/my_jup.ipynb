# Life cycle of creating plot using matplotlib


```python
fig,(rome1,rome2) = plt.subplots(2,1,figsize=(20,10),sharex=True) #making canvus with 2 rows and 1 column
scatter_plot = rome1.scatter(       #adding data to canvus
    x=df['new_clm_2'],
    y=df['cuml_cml'],
    c=df['target'],
    cmap='viridis', #adding colour
)
plt.legend(*scatter_plot.legend_elements(),  #adding legend(title of susceplibility) to scatter plot
           title='Target')
scatter_plot2 = rome2.scatter(     #adding data to canvus
    x=df['age'],
    y=df['cuml_cml'],
    c=df['target'],
    cmap='plasma',
)
plt.legend(*scatter_plot2.legend_elements(),  #adding legend(title of susceplibility) to scatter plot
              title='Target')
rome1.set(                                     # customizing rome1 of its lables and title
    title='Age vs Cholesterol',
    #xlabel='Age', #because x label is shared
    ylabel='Cholesterol',
)
rome2.set(                                     # customizing rome2 of its lables and title
    title='Age vs Max Heart Rate',
    xlabel='Age',
    ylabel='Max Heart Rate',
)
rome1.axhline(y=df['chol'].median(),color='r',linestyle='--') #adding mean and median lines to the plots
rome2.axhline(y=df['thalach'].mean(),color='r',linestyle='--')
rome1.axhline(y=df['chol'].mean(),color='g',linestyle='-')  #adding mean and median lines to the plots
rome2.axhline(y=df['thalach'].median(),color='g',linestyle='-')
fig.suptitle('Heart Disease Analysis',fontsize=32,fontweight='bold') #adding overall title to the canvus
plt.style.use( 'seaborn-v0_8-pastel') #applying style to the canvus
fig.savefig('heart_disease_analysis.png') #saving the canvus as a png file
plt.show() #displaying the canvus
```
